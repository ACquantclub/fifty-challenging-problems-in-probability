- In an election, two candidates, Albert and Benjamin, have in a ballot box $a$ and $b$ votes respectively with $a>b$
- Ballots are randomly drawn and tallied
- *What is the chance that at least once after the first tally, the candidates have the same number of tallies?*

____

- First, before we start solving this problem, let's simulate this process

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
plt.rcParams['figure.figsize'] = 10, 10
```

```python
list_final_results = []
n_trials = 2000

for a in range(2, 100):
    for b in range(1, a):
        count = 0
        array_votes = np.array(a*[1]+b*[-1])
        for trial in range(n_trials):
            np.random.shuffle(array_votes)
            array_tally = np.cumsum(array_votes)
            if 0 in array_tally:
                count += 1
        list_final_results.append((a, b, count/n_trials))
```

```python
x = [a for a, b, c in list_final_results]
y = [b for a, b, c in list_final_results]
color = [c for a, b, c in list_final_results]
```

```python
plt.scatter(x, y, c=color)
plt.colorbar()
plt.xlabel('a (Votes for Albert)')
plt.ylabel('b (Votes for Benjamin)')
plt.show()
```

```python
df_summary = pd.DataFrame({'a':x, 'b':y, 'P':color}).pivot(index='a', columns='b',values='P')
```

```python
df_summary_sample = df_summary.iloc[::10].T
df_summary_sample.plot()
```

- As we can see from the plot, **the closer a and b are, the higher the probability that the tallies will be equal at some point in the counting**

_____

- To solve this problem, we want to calculate the probability that they never have the same number of tallies
    - Then, we take one minus our answer to get the solution

- If the first tally is for Bejamin, then it's **impossible that they never have the same number of tallies**
    - *Why?*
        - Imagine Benjamin has 2 votes in the box, and Albert has 3
            - Then, no matter what, at some point they'll run out votes to count for Benjamin and will need to keep counting Albert's
                - And since he has more, we know they have to cross at some point
    - Another way to think about it is if we have a string that connects from (0, 1) to (a+b, a) and a second that connects from (0,0) to (a+b, a) (where a>b as described), then the strings have to cross at some point
        - This is like the [Intermediate Value Theorem](https://en.wikipedia.org/wiki/Intermediate_value_theorem)

- Therefore, if we draw a vote for Benjamin first, the probability that the candidates have the same number of tallies at some point is 1

- *But what if we draw a vote for Albert first?*
    - For the tallies to never be equal, we always need to have more Albert votes than Benjamin
        - This has non-zero probability

- If the first vote goes to Albert, and we want to consider only scenarios where Benjamin doesn't catch up, then we need the second vote to also go to Albert
    - Otherwise they're tied

- So, **all scenarios where they're never tied in tallies start with the first two votes going to Albert**

- If Albert has a two point lead, then for Benjamin to catch up in the next two draws, **he needs to get two votes in a row**
    - If it's one and one, they cancel eachother out and the vote gap remains at 2

$$
P(\text{Benjamin gets 2 in a row | First 2 go to Albert}) = \frac{b}{a+b-2}\frac{b-1}{a+b-3}
$$

- So, the probability that Albert is still in the lead after 4 draws given he got the first two is:

$$
P(\text{Albert leading after 4 draws | Albert leading after 2 draws}) = 1 - \left (\frac{b}{a+b-2}\frac{b-1}{a+b-3} \right )
$$

- But now, Albert isn't just ahead by just 2 in each scenario anymore
    - In one of the paths, both wins went to Albert, so now he's leading by 4 instead of 2
