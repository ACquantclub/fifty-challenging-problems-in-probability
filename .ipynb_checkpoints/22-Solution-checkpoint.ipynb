- In an election, two candidates, Albert and Benjamin, have in a ballot box $a$ and $b$ votes respectively with $a>b$
- Ballots are randomly drawn and tallied
- *What is the chance that at least once after the first tally, the candidates have the same number of tallies?*

____

- First, before we start solving this problem, let's simulate this process

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
plt.rcParams['figure.figsize'] = 10, 10
```

```python
list_final_results = []
n_trials = 10000

for a in range(2, 100):
    for b in range(1, a):
        count = 0
        array_votes = np.repeat(np.array(a*[1]+b*[-1])[:,np.newaxis], n_trials, axis=1)
        idx_random = np.random.rand(*array_votes.shape).argsort(0)
        array_votes_shuffled = array_votes[idx_random, np.arange(array_votes.shape[1])]
        array_tally = np.cumsum(array_votes_shuffled, axis=0)
        array_ties = (np.amin(array_tally, axis=0)<=0).astype(int)
        list_final_results.append((a, b, np.mean(array_ties)))
```

```python
x = [a for a, b, c in list_final_results]
y = [b for a, b, c in list_final_results]
color = [c for a, b, c in list_final_results]
```

```python
plt.scatter(x, y, c=color)
plt.colorbar()
plt.xlabel('a (Votes for Albert)')
plt.ylabel('b (Votes for Benjamin)')
plt.show()
```

```python
df_summary = pd.DataFrame({'a':x, 'b':y, 'P':color}).pivot(index='a', columns='b',values='P')
```

```python
df_summary_sample = df_summary.iloc[::10].T
df_summary_sample.plot()
plt.ylabel('Probability')
plt.legend(loc='lower right')
plt.show()
```

- As we can see from the plot, **the closer a and b are, the higher the probability that the tallies will be equal at some point in the counting**

_____

- To get to the solution, let's start with a simple case
    - $a=2, b=1$
    
- So, there are 3 ways the votes can be tallied
    1. [A, A, B]
    2. [A, B, A]
    3. [B, A, A]
    
- In scenario 1, they never have the same number of votes
    - In scenarios 2 and 3, there are moments where they are tied in votes
    
- **Therefore, the probability that their tallies will be equal is 2/3**

- As we can see from this example, the number of paths can be counted by thinking about the number of ways we can place the votes for Benjamin
    - There are 3 votes total to be tallied, and one of them must go to Benjamin
        - Therefore there are $\binom{3}{1}$ different ways to tally the votes

- So, in general, there are $\binom{a+b}{b}$ different ways to tally the votes
    - **Note**: since $\binom{a+b}{b} = \frac{(a+b)!}{(a+b-b)!b!} = \frac{(a+b)!}{a!b!} = \frac{(a+b)!}{a!(a+b-a)!} = \binom{a+b}{a}$, we could have used $\binom{a+b}{a}$, but it doesn't affect our calculations

- Now, let's think about a bigger example
    - $a = 4, b = 2$
        - There are $\binom{4+2}{2} = 15$ different possible paths

```python
from itertools import combinations
```

```python
df_example = pd.DataFrame('a', columns=range(15), index=range(6))

for i, indices in enumerate(combinations(range(6), 2)):
    df_example.loc[indices, i] = 'b'
```

```python
df_example
```

- Now, we map these orders and convert them to tallies

```python
df_tally = ((df_example=='a')*1 + (df_example=='b')*(-1)).cumsum()
```

```python
df_tally
```

- So, each column represents a possible path, and when the tally is equal to zero, the vote count is equal

```python
df_tally.plot()
plt.axhline(0, ls='dotted', c='r')
```

- As we can see, any of the paths that touch the dotted line have at least one tie at some point
    - Let's look only at the paths where there is **no tie**

```python
df_tally[[x for x in df_tally.columns if df_tally[x].min()>0]].plot()
plt.axhline(0, ls='dotted', c='r')
```

- As we can see, in all cases, Albert gets off to a 2 vote head start

- If the first tally is for Bejamin, then it's **impossible that they never have the same number of tallies**
    - *Why?*
        - If we have a string that connects from (0, 1) to (a+b, a) and a second that connects from (0,0) to (a+b, a) (where a>b as described), then the strings have to cross at some point
            - This is like the [Intermediate Value Theorem](https://en.wikipedia.org/wiki/Intermediate_value_theorem)

- Therefore, if we draw a vote for Benjamin first, the probability that the candidates have the same number of tallies at some point is 1

- If the first vote goes to Albert, and we want to consider only scenarios where Benjamin doesn't catch up, then we need the second vote to also go to Albert
    - Otherwise they're tied

- So, **all scenarios where they're never tied in tallies start with the first two votes going to Albert**

- Now, let's look at a few other examples and count the number of paths where they're never tied

```python
from scipy.special import comb
```

```python
for a in range(2, 8):
    for b in range(1,a):
        n_votes = a + b
        n_paths = int(comb(n_votes, b))
        df_example = pd.DataFrame(1, columns=range(n_paths), index=range(n_votes))

        for i, indices in enumerate(combinations(range(n_votes), b)):
            df_example.loc[indices, i] = -1
            
        df_tally = df_example.cumsum()
        
        count = len([x for x in df_tally.columns if df_tally[x].min()>0])
        print('a = {}, b = {}, P = {}/{} = {}'.format(a, b, count, n_paths, count/n_paths))
```
